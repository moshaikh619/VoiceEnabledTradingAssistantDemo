```python
# testing gesture control

from cvzone.HandTrackingModule import HandDetector
import cv2
import os
import numpy as np
import pyttsx3

tts = pyttsx3.init()

#  speaking and print 
def speak_text(text):
    print(text)
    tts.say(text)
    tts.runAndWait()


print("Starting gesture recognition...")


width, height = 1000, 800
gestureThreshold = 300
# Camera Setup
cam = cv2.VideoCapture(0)
cam.set(3, width)
cam.set(4, height) 
# Hand Detector
detectorHand = HandDetector(detectionCon=0.8, maxHands=1)
# Variables
delay = 30
buttonPressed = False
counter = 0


while True:
    # Get image frame
    success, img = cam.read()
    # Find the hand and its landmarks
    hands, img = detectorHand.findHands(img)  # with draw
    if hands and buttonPressed is False:  # If hand is detected
        hand = hands[0]
        cx, cy = hand["center"]
        lmList = hand["lmList"]  # List of 21 Landmark points
        fin = detectorHand.fingersUp(hand)  
        if cy <= gestureThreshold:  # If hand is at the height of the face
           
            if fin == [1, 1, 1, 1, 1]: 
                #print("Next")
                speak_text("Next")
                buttonPressed = True
               
            if fin == [1, 0, 0, 0, 0]:
                #print("Previous")
                speak_text("Previous")
                buttonPressed = True


            if fin == [0, 1, 1, 0, 0]: 
                
                speak_text("increasing brightness")
                from screen_brightness_control import set_brightness, get_brightness

                current_brightness = 0
                new_brightness = 0

                # //current_brightness = get_brightness()

                # print(current_brightness)

                # # new brightness level
                # new_brightness = current_brightness + 10 
                # print(new_brightness)
                # # replace this value with the desired brightness level
                # if new_brightness <= 100 :
                #     set_brightness(new_brightness)
                # else: 
                new_brightness = 100
                # display the current brightness level
                current_brightness = get_brightness()
                set_brightness(new_brightness)
                print("Current screen brightness is:", current_brightness)
                buttonPressed = True
               
            if fin == [0, 0, 0, 0, 0]:
                
                speak_text("decreasing brightness")

                from screen_brightness_control import set_brightness, get_brightness

                # current_brightness = 0
                # new_brightness = 0
                
                # current_brightness = get_brightness()

                # new_brightness = current_brightness - 10
                # # new brightness level
                # if new_brightness >= 0 :
                #     set_brightness(new_brightness)
                # else:
                new_brightness = 10
               

                # replace this value with the desired brightness level
                set_brightness(new_brightness)
                # display the current brightness level
                current_brightness = get_brightness()
                print("Current screen brightness is:", current_brightness)  
                buttonPressed = True


            if fin == [0,1,1,1,0]:

                speak_text("opening powerpoint")
                
                import win32com.client as win32
                import time

                app = win32.Dispatch("PowerPoint.Application")

                objCOM = app.Presentations.Open(FileName="Project Inclusify.pptx", WithWindow=1)

                #START THE SLIDESHOW
                objCOM.SlideShowSettings.Run()

                time.sleep(2.5)
                #MOVE TO THE NEXT SLIDE
                objCOM.SlideShowWindow.View.Next()
                buttonPressed = True
              
 
    if buttonPressed:
        counter += 1
        if counter > delay:
            counter = 0
            buttonPressed = False

 
    cv2.imshow("GestureRecognition", img)
 
    key = cv2.waitKey(1)
    if key == ord('q'): # press q to quit
        cv2.destroyAllWindows()
        break

```
